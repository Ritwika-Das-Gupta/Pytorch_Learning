## Tensors
- A torch, tensor is a multi - dimensional matrix containing elements of a sgle data type
- Similar to numpy arrays, but full of things that will make them work better GPUs (vs regular CPU)
- default data type of float 32
- More suitable for deep learning than a numpy array
